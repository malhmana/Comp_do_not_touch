```python
# Reading the CSV file
df= pd.read_csv(r'C:\Users\Work\Downloads\UPD_ Forbidden companies Clients as Do not touch- preliminary - UPD_ Forbidden companies Clients as Do not touch- preliminary.csv')
```

```python
## Creating a list of clients to check with current company
forbidden_comp= list(df['Unnamed: 6'].dropna())
```

```python
len(forbidden_comp)
```

```python
df.head()
```

```python
## What this codeblock does: It goes through all the rows and checks if Current Company has a forbidden company in it.
### Then it adds that company name to Company do not touch column.


x= 0 # Index
while x<len(df):
    for i in forbidden_comp:
        if i.lower() in df.iloc[x]['Current Company'].lower():
            df['Company do not touch'].iloc[x]= i
    x += 1 # Index + 1 for while loop to end
```

```python
# This is an idea code, i.e. it will not work for our situation because some people name their company differently on LI.

for i in range (len(df)):
    comp_names= df.iloc[i]['Current Company'].split(";")
    comp_names_lower= [compa.lower() for compa in comp_names]
    for comp in forbidden_comp:
        if comp.lower() in comp_names_lower:
            df['Company do not touch'].iloc[i]= comp
                
```
